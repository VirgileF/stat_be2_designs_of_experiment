### Questions 1 et 2

```r
hormones = read.table('hormones.txt', header=T, dec=',')
summary(hormones)
hormones
```

```r
hormones$X1 = (hormones$X1 - 85) / 35
hormones$X2 = (hormones$X2 - 20) / 5
hormones
```

```r
plot(hormones$X1, hormones$X2)
```

Les 4 premiers points de mesure constitue un plan factoriel complet $2^2$. Les autres points de mesure vont permettre d'obtenir un modèle plus performant. Le plan d'expérience permet aussi d'estimer l'influence de l'intéraction de $X_1$ et $X_2$.

```r
hormones$X1X2 = hormones$X1 * hormones$X2
```

### Question 3

```r
# Calcul de la matrice (tXX)^(-1)
X = data.matrix(hormones[,1:])
solve(t(X)%*%X)
```

On voudrait approcher au mieux les fonctions $f_1:(X_1,X_2)\mapsto Y_1$ et $f_2:(X_1,X_2)\mapsto Y_2$. Toutes les variables sont continues, on va donc privilégier un modèle linéaire. La matrice $^tXX$ est inversible donc ce plan permet d'estimer un tel modèle.

### Question 4

```r
# Premier groupe d'enfants
lm1 = lm(Y1 ~ X1 + X2 + X1X2, data=hormones)
summary(lm1)
plot(lm1)
```

L\'intéraction $X_1 X_2$ ne semble pas influente car la *p-value* associée est supérieure à 0.05. On privilégera donc un modèle sans intéraction. 

```r
# Premier groupe d'enfants sans intéraction
lm2 = lm(Y1 ~ X1 + X2, data=hormones)
summary(lm2)
plot(lm2)
```

On observe que la 9ème valeur est aberrante, on peut essayer un modèle sans en prendre compte. 

```r
# Premier groupe d'enfants sans intéraction et sans l'outlier
hormones_minus_outlier = hormones[1:8,]
hormones_minus_outlier
lm3 = lm(Y1 ~ X1 + X2, data=hormones_minus_outlier)
summary(lm3)
plot(lm3)
```

On gardera le modèle sans interaction $(X_1, X_2) \mapsto Y_1$ après avoir retiré la valeur aberrante du jeu de données, car c'est celui qui permet d'expliquer le plus de variance ($R^2=0.9675$).

### Question 5

Il s'agit d'une régression linéaire à coefficients positifs donc on la fonction est strictement croissante et non bornée, on ne peut donc pas maximiser la croissance. D'autre part, ça n'a pas de sens de choisir un traîtement de sorte à maximiser la croissance, on tentera plutôt de la rendre la plus proche possible de la moyenne de la population (une croissance "normale"). D'après les courbes d'IMC (https://www.magicfoundation.org/downloads/boysgrowthchartCDC240.pdf pour les garçons et https://www.magicfoundation.org/downloads/girlsgrowthchartCDC240.pdf pour les filles), un enfant, indépendamment de son sexe, mesure en moyenne 102 cm à 4 ans et 115 à 6 ans donc un rythme de croissance normal serait à peu près 3.75 cm en 6 mois. Le problème revient donc à trouver des valeurs $x_1$ et $x_2$ telles que $12.1450 + 1.9606\times x_1 + 2.7862\times x_2 = 3.75$. Cela revient à s'assurer que $x_2 = -3.01 - 0.7 \times x_1$. Faisons en sorte de minimiser les prises d'hormones afin de limiter les effets secondaires. $x_1=x_2=-1.75$ paraît être un bon compromis. On pensera à remettre à l'échelle pour obtenir les fréquences des traîtements en ppm : on obtient donc $X_1 = 23.75 ppm$ et $X_2 = 11.25 ppm$.

```r
newdata = data.frame(X1=(100-85)/35, X2=(100-20)/5)
predict(lm3, newdata=newdata, interval=c("confidence"), level = 0.95)
```

On obtient une valeur extrêmement grande, ce qui est logique car les prises d'hormones sont beaucoup plus fréquentes que la solution optimale trouvée.

### Question 6

```r
# Deuxième groupe d'enfants
lm4 = lm(Y2 ~ X1 + X2 + X1X2, data=hormones)
summary(lm4)
plot(lm4)
```

```r
# Deuxième groupe d'enfants sans intéraction
lm5 = lm(Y2 ~ X1 + X2, data=hormones)
summary(lm5)
plot(lm5)
```

```r
# Deuxième groupe d'enfants sans X2
lm6 = lm(Y2 ~ X1, data=hormones)
summary(lm6)
plot(lm6)
```

```r
# Deuxième groupe d'enfants avec estimation quadratique
hormones$X1sq = hormones$X1^2
hormones$X2sq = hormones$X2^2
hormones
lm7 = lm(Y2 ~ X1 + X2 + X1sq + X2sq, data=hormones)
summary(lm7)
plot(lm7)
```

C'est ce modèle qui permet d'expliquer le plus de variance, on le choisira donc pour la recherche de solution optimale. Avec une telle estimation quadratique, on peut maximiser la croissance (car la fonction tend vers -inf lorsque x1 et x2 tendent vers +inf), on va donc pouvoir répondre à la question. Il faut trouver les valeurs de $x_1$ et $x_2$ telles que $c(x_1, x_2) = \beta_0 + \beta_1 x_1 + \beta_2 x_2 + \beta_3 x_1^2 + \beta_4 x_2^2$ soit maximale. Or $\frac{\partial c}{\partial x_1} = \beta_1 + 2\beta_3 x_1$ et $\frac{\partial c}{\partial x_2} = \beta_1 + 2\beta_3 x_2$, ainsi on trouve la solution optimale $x_1^*=0.24, x_2^*=0.61$. 

#### Ré-appliquons le modèle quadratique pour le premier groupe :

```r
# Premier groupe d'enfants sans intéraction et avec estimation quadratique
hormones$X1sq = hormones$X1^2
hormones$X2sq = hormones$X2^2
hormones
lm8 = lm(Y1 ~ X1 + X2 + X1sq + X2sq, data=hormones)
summary(lm8)
plot(lm8)
```

De même, il faut trouver les valeurs de $x_1$ et $x_2$ telles que $c(x_1, x_2) = \beta_0 + \beta_1 x_1 + \beta_2 x_2 + \beta_3 x_1^2 + \beta_4 x_2^2$ soit maximale. Or $\frac{\partial c}{\partial x_1} = \beta_1 + 2\beta_3 x_1$ et $\frac{\partial c}{\partial x_2} = \beta_1 + 2\beta_3 x_2$, ainsi on trouve la solution optimale $x_1^*=0.49, x_2^*=0.87$.

#### Modèle quadratique sur les deux groupes

```r
hormones_group1 = hormones
hormones_group2 = hormones
hormones_group1$Y = hormones_group1$Y1
hormones_group2$Y = hormones_group2$Y2
hormones_final = rbind(hormones_group1, hormones_group2)
hormones_final
# La colonne Y doit être égale à Y1 sur les 9 premières lignes et Y2 sur les 9 suivantes.
```

```r
# Tous les enfants sans intéraction et avec estimation quadratique
hormones_final$X1sq = hormones_final$X1^2
hormones_final$X2sq = hormones_final$X2^2
lm9 = lm(Y ~ X1 + X2 + X1sq + X2sq, data=hormones_final)
summary(lm9)
plot(lm9)
```

Le résultat n'est pas mieux en terme de variance expliquée.
