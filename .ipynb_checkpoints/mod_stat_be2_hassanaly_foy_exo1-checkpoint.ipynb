### Question 1

```r
silicium = read.table("silicium.txt", header=TRUE)
summary(silicium)
```

```r
silicium[,1:7]
nrow(silicium)
```

```r
# Calcul de la matrice (tXX)^(-1)
X = data.matrix(silicium[,1:6])
solve(t(X)%*%X)
1/64
```

```r
silicium$ABC = silicium$Ltemp * silicium$Ltime * silicium$Lpress
silicium$ACD = silicium$Ltemp * silicium$Lpress * silicium$Ctemp
identical(silicium$ABC, silicium$Ctime)
identical(silicium$ACD, silicium$Catmos)
```

- $(^tXX)^{-1}$ est un multiple de la matrice Id donc les facteurs sont orthogonaux.
- On remarque que Ctime = Ltemp * Ltime * Lpress (E = ABC) et Catmos = Ltemp * Lpress * Ctemp (F = ACD) donc il s'agit du plan fractionnaire $2^{6-2}$ répété 4 fois. Il est de résolution $IV$.
- Ce plan permet d'estimer les effets principaux mais pas les intéractions doubles (il est donc sans confusion). On a probablement privilégié le coût de mise en oeuvre en divisant par 4 le nombre de points de mesure, mais on a répété chaque mesure 4 fois afin de diminuer la variance de l'estimateur (on trouve 1/64 au lieu de 1/16 avec seulement 16 mesures). 

### Questions 2 et 3

```r
lm1 = lm(Camber ~ Ltemp + Ltime + Lpress + Ctemp + Ctime + Catmos, data=silicium)
summary(lm1)
```

```r
anova(lm1, test='Chisq')
```

```r
38.34/sqrt(64)
```

- La p-value globale est bien inférieure à .05 donc il y a au moins une variable significative.
- Cependant, *adjusted R^2* = 0.67 donc le modèle à 6 variables ne permet d'expliquer que les 2 tiers de la variance totale, ce quiest un bon début, mais pas satisfaisant.
- Chaque estimateur $\widehat{\beta}_k$ suit une loi normale de moyenne $\beta_k$ et de d'écart-type $\widehat{\sigma_k}=\widehat{\sigma}\sqrt{(^tXX)^{-1}_{(i,i)}}$ où $\widehat{\sigma}=RSE$. Ainsi on doit trouver $\widehat{\sigma_k}=\frac{38.34}{\sqrt{64}}=4.793$. 
- On observe aussi que les facteurs Ltime et Ctemp ont des *p-values* supérieures à .05. Ainsi, ces variables ne semblent pas explicatives dans le modèle ainsi défini.
- Avec la fonction *anova* on retrouve bien que Ltime et Ctemp ne permettent pas d'expliquer une grande partie de la variance du modèle.

### Question 4

```r
lm2 = lm(Camber ~ Ltemp + Lpress + Ctime + Catmos, data=silicium)
summary(lm2)
```

- On obtient les mêmes performances avec ce modèle, le R^2 n'a pratiquement pas diminué. On privilégiera ce dernier.
- Les coefficients associés aux variables sélectionnées sont conservés.

### Question 5

```r
newdata = data.frame(Ltemp=-1, Lpress=-1, Ctime=1, Catmos=1)
predict(lm2, newdata=newdata, interval=c("confidence"), level = 0.95)
```

```r
107.016 - 19.453 - 28.016 -17.234 - 38.734 + 5 * 4.816
```

- Limiter la courbure au maximum revient à minimiser la fonction estimée (fonction affine en les variables d'entrée). Les variables prennent leurs valeurs dans {-1, 1} donc on prendra Ltemp=Lpress=-1 (car leurs coefficients associés sont positifs) et Ctime=Catmos=1 (car leurs coefficients associés sont négatifs). Ltime et Ctemp ne sont pas significatives donc on cherchera à minimiser le coût d'énergie, ainsi on prendre Ltime=Ctemp=-1. En résumé, les meilleures conditions de fabrication sont:
    - Durée de laminage : 10 sec
    - Température du laminage : 55 deg
    - Pression du laminage : 5 bar
    - Température de cuisson : 1580 deg
    - Temps de cuisson : 29 sec
    - Température de rosée de cuisson : 26 deg
- L'intervalle de confiance pour ces valeurs de paramètres est \[-17.97, 25.13\]. 
- Une augmentation de la température de laminage de 5 degrés provoquera une augmentation de Ltemp de 1/2 (le calcul est $(+1 - (-1))\frac{60-55}{75-55}$), donc une augmentation de la courbure de 1/2\*19.453=9.727.

```r
# Plot des erreurs de prédiction
# newdata2 = silicium[,1:7]
# silicium$predicted_Camber = predict(lm2, newdata=newdata2, interval=c("confidence"), level = 0.95)[,1]
# silicium$pred_error = silicium$predicted_Camber - silicium$Camber
# plot(silicium$pred_error, type='l')

# QQ-plots
plot(lm2)

# Covariances des variables avec les erreurs
cov(silicium$pred_error, silicium$Ltemp)
cov(silicium$pred_error, silicium$Lpress)
cov(silicium$pred_error, silicium$Ctime)
cov(silicium$pred_error, silicium$Catmos)
```

- **Indépendance des variables explicatives :** $^tXX$ est inversible donc les variables sont indépendantes les unes des autres.
- **Indépendance des erreurs :** La courbe des résidus n'est pas constante donc il n'y a pas une parfaite décorrélation des erreurs. L'hypothèse n'est donc pas rigoureusement vérifiée.
- **Exogénéité :** Les calculs des covariances de chaque variable avec les erreurs donnent des valeurs très proches de 0, ce qui laisse penser que les erreurs sont bien indépendantes des variables explicatives.
- **Homoscédasticité :** Le nuage de points obtenus dans le graphe *Residuals vs Fitted* n'est pas parfaitement homogène donc l'homoscédasticité n'est pas rigoureusement vérifiée.
